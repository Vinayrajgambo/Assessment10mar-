```python
#Second python question
def ok(v,x,c):
    n=len(v)
    count=1
    d=v[0]
    for i in range(1,n):
        if(v[i]-d>=x):
            d=v[i]
            count=count+1
        else:
            continue
    if(count>=c):
        return True
     
    return False
def aggressive_cows(v,n,k):
    ans=-1
    maxi=0
    for i in range(n):
        maxi=max(maxi,v[i]) 
    for i in range(1,maxi+1):
        if(ok(v,i,k)):
            ans=i
        else:
            break
    return ans
K=3
arr=[1,2,4,8,9]
N=len(arr)
ans=aggressive_cows(arr,N,K)
print(ans)
```

```python
#4th python answer
class Pair:
    def __init__(self, x, y):
        self.index1 = x
        self.index2 = y
def GetQuadruplets(nums, target):
    map = {}
    for i in range(len(nums) - 1):
        for j in range(i + 1, len(nums)):
            sum = nums[i] + nums[j]
            if sum not in map:
                map[sum] = [Pair(i, j)]
            else:
                map[sum].append(Pair(i, j))
    ans = set()
    for i in range(len(nums) - 1):
        for j in range(i + 1, len(nums)):
            lookUp = target - (nums[i] + nums[j])
            if lookUp in map:
                temp = map[lookUp]
                for pair in temp:
                    if pair.index1 != i and pair.index1 != j and pair.index2 != i and pair.index2 != j:
                        l1 = [nums[pair.index1], nums[pair.index2], nums[i], nums[j]]
                        l1.sort()
                        ans.add(tuple(l1))
    print(*reversed(list(ans)), sep = '\n')

arr = [1, 0, -1, 0, -2, 2]
K = 0
GetQuadruplets(arr, K)
```

```python
#1st answer Statictics
'''
Six Sigma is a methodology used to improve the quality of processes by minimizing the number of defects or errors. 
It relies on statistical analysis to measure and control variation in processes. 

Meaning in Statistics:

Six Sigma refers to a specific level of quality achievable in a process. 
It indicates a very high process capability, meaning the process can produce near-flawless outputs with a very low defect rate.
Statistically, Six Sigma processes are capable of producing only 3.4 defects per million opportunities (DPMO). 
This translates to a 99.99966% success rate.
It's achieved by focusing on minimizing variation in the process. Variation is any deviation from the desired outcome, and it can lead to defects.
Example:
In a Six Sigma scenario:

Mean (μ): The target length is 10 centimeters.
Standard Deviation (σ): Let's assume a standard deviation of 0.2 centimeters, indicating the natural variation in widget length.
A Six Sigma process would have a performance level such that the specification limits (acceptable range) are within six standard deviations from the mean. 
For this example, the acceptable range would be from 
10−6×0.2 to 10+6×0.2, or 9.2 to 10.8 centimeters.

Imagine a manufacturing company that produces bottles.

A process without Six Sigma might have a defect rate of, say, 1%. This means 1 out of every 100 bottles might be defective (cracked, misshapen, etc.).
By implementing Six Sigma principles and analyzing the bottle-making process, the company can identify and address factors that cause defects. 
This could involve tightening quality control measures, improving machine calibration, or optimizing material usage.
With Six Sigma in place, the defect rate can be significantly reduced, leading to a much smaller number of defective bottles (e.g., only 3.4 per million produced).
Benefits of Six Sigma:

Reduced costs: Fewer defects lead to less wasted material, lower rework costs, and improved customer satisfaction.
Increased efficiency: Streamlined processes with minimal variation lead to faster production times and better resource utilization.
Enhanced customer satisfaction: Consistent high-quality products lead to happier customers and improved brand reputation.
Six Sigma is a powerful tool used in various industries, including manufacturing, healthcare, finance, and service sectors. 
It helps organizations achieve operational excellence by focusing on continuous improvement and minimizing defects.'''
```

```python
#3rd answer statistical

Minimum: The smallest value in the dataset.
First Quartile (Q1): The value below which 25% of the data falls. It is the median of the lower half of the dataset.
Median (Q2): The middle value of the dataset. It divides the dataset into two halves, with 50% of the data below and 50% above.
Third Quartile (Q3): The value below which 75% of the data falls. It is the median of the upper half of the dataset.
Maximum: The largest value in the dataset.

example
{12,15,18,21,24,25,27,30,32,35,40,50,60}

Minimum: 12
First Quartile (Q1): 20.5 (average of the 6th and 7th values)
Median (Q2): 27
Third Quartile (Q3): 35 (average of the 11th and 12th values)
Maximum: 60
```

```python
#4th answer
'''
Positive correlation (1): As one variable increases, the other variable also increases.
Negative correlation (-1): As one variable increases, the other variable decreases.
No correlation (0): There is no linear relationship between the variables. '''


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(42)
x = np.random.rand(50) * 10   
y = 2 * x + 1 + np.random.randn(50) * 2   
df = pd.DataFrame({'X': x, 'Y': y})
correlation_coefficient = df['X'].corr(df['Y'])
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=df, color='blue')
plt.title(f'Scatter Plot with Correlation Coefficient: {correlation_coefficient:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
